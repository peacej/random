```python
import os, json
import pandas as pd
```

```python
json_file = '~/Downloads/highmonks_metadata.json'
```

```python
df = pd.read_json(json_file)
```

```python
df = df.explode('attributes')
```

```python
df = df.reset_index(drop=True)
```

```python
NUM_MONKS = 11111
```

```python
assert NUM_MONKS == len(df['dna'].unique())
```

```python
df.head()
```

```python
df['attributes'][0]
```

```python
def grab_trait(x):
    return x['trait_type']
```

```python
def grab_value(x):
    return x['value']
```

```python
def grab_rarity(x):
    return x.split(' ')[-1]
```

```python
df['trait'] = df['attributes'].apply(lambda x: grab_trait(x))
df['value'] = df['attributes'].apply(lambda x: grab_value(x))
df['rarity'] = df['value'].apply(lambda x: grab_rarity(x))
```

```python
sum(df['value'].str.contains('GOLDEN JOINT'))
```

```python
for rarity in ['Common','Rare','Legendary','Ancient']:
    print(rarity, ': ', sum(df['value'].str.contains(rarity)) / (NUM_MONKS*9))
```

```python
# verify 9 attributes per monk
sum(df.groupby('dna').size() != 9)
```
